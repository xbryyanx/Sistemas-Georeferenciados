Que es un GeoDataFrame?
Un objeto GeoDataFrame es una estructura de datos tabular que contiene por lo menos una columna con un objeto (Geoseries) que contiene la geometría.

Que es Series?
Series
Una "Series" es una estructura de datos unidimensional que puede contener cualquier tipo de datos (enteros, cadenas, flotantes, objetos de Python, etc.), similar a una columna en una hoja de cálculo o una tabla de base de datos. Es proporcionada por la biblioteca pandas, que es ampliamente utilizada para la manipulación y el análisis de datos en Python.

Las Series son básicamente un arreglo de datos etiquetado, lo que significa que cada elemento en la Series tiene un índice asociado. Este índice puede ser un número (índice por defecto), una cadena (etiqueta), o incluso fechas (DateTime).

Funcionamiento de Series:

Creación: Puedes crear una Series a partir de listas, diccionarios, o arreglos de numpy.
Manipulación: Una vez creada, puedes realizar operaciones sobre ella como sumas, medias, filtrados, entre otros, de manera muy eficiente.
Indexación: Puedes acceder a los elementos de una Series usando su índice, lo cual permite recuperar o modificar datos de manera muy conveniente.


Que es GeoSeries?

GeoSeries
Un "GeoSeries" es una estructura de datos especializada, similar a las Series de pandas, pero diseñada específicamente para almacenar datos geométricos (puntos, líneas, polígonos). Es proporcionada por la biblioteca geopandas, que extiende las funcionalidades de pandas para permitir el manejo eficiente de datos geoespaciales.

Una GeoSeries puede considerarse como una Series donde cada elemento es un objeto geométrico (representado usando el tipo de datos shapely). Esto significa que cada entrada en una GeoSeries puede ser, por ejemplo, una representación de una ubicación en la tierra, un camino, o una región.

Funcionamiento de GeoSeries:

Creación: Al igual que con pandas, puedes crear una GeoSeries a partir de una lista de objetos geométricos.
Operaciones geoespaciales: GeoSeries permite realizar operaciones específicas de datos geoespaciales, como calcular áreas, longitudes, intersecciones, uniones, y mucho más.
Visualización: GeoPandas facilita la visualización de datos geoespaciales, permitiendo un análisis visual directo de la información geográfica.

Para que sirven?

Para qué sirven
Series: Sirven para manejar y analizar datos unidimensionales de manera eficiente. Son útiles en todo tipo de análisis de datos donde la manipulación y transformación de series temporales, listas de elementos, o cualquier conjunto de datos unidimensional es necesaria.
GeoSeries: Están diseñadas para el manejo y análisis de datos geoespaciales. Son esenciales en el campo de la geografía, planificación urbana, navegación, y ciencias ambientales, donde la manipulación y análisis de datos geográficos son cruciales.

Como funcionan?

Las Series, proporcionadas por pandas, son útiles en una amplia gama de aplicaciones de análisis de datos debido a su flexibilidad y eficiencia para manejar datos unidimensionales.

Las GeoSeries, por otro lado, son proporcionadas por GeoPandas y están diseñadas específicamente para trabajar con datos geoespaciales. Son cruciales en campos como la geografía, planificación urbana, y ciencias ambientales.

Pregunta 5. ¿Para qué sirve plot? ¿Qué es markersize?
Claro, con gusto. En programación, particularmente en Python utilizando bibliotecas como Matplotlib, la función plot() se emplea para visualizar datos en forma de gráficos. Puede utilizarse para crear gráficos de líneas, gráficos de dispersión, histogramas y más. Esta función toma datos numéricos y los representa visualmente en un gráfico, permitiendo así analizar y comprender mejor los patrones, tendencias o distribuciones presentes en los datos.

Por otro lado, markersize es un parámetro específico de la función plot() que se utiliza para controlar el tamaño de los marcadores en un gráfico, como puntos individuales en un gráfico de dispersión. Los marcadores son símbolos visuales que representan cada punto de datos en el gráfico. Al ajustar el parámetro markersize, puedes controlar el tamaño de estos marcadores para que sean más grandes o más pequeños, según tus preferencias o necesidades de visualización. Esto es útil para resaltar ciertos puntos o ajustar la estética del gráfico para que se adapte mejor a tus requerimientos de presentación o análisis.

```python
import matplotlib.pyplot as plt
colombia.plot(markersize=0.5);
```

Pregunta 7. ¿Para que se usan los centroides en los SIG?
Los centroides en los Sistemas de Información Geográfica (SIG) son puntos que representan el centro geométrico de polígonos. Se utilizan para simplificar la representación de áreas, realizar análisis espaciales, asociar datos tabulares con geometría espacial y en operaciones de geoprocesamiento.

```python
# ahora creas centroides y lo conviertes en la geometria:

colombia['centroid_column'] = colombia.centroid
colombia = colombia.set_geometry('centroid_column')
colombia.plot() 
```
